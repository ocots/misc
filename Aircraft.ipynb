## Aircraft trajectories in climbing phase and singular perturbations


We are interested in the time optimal control of an aircraft during its climbing phase. 


This phase is determined by its own dynamics governed by the fourth dimensional system

$$
\frac{dh}{dt} = v sin(\gamma) \\
\frac{dv}{dt} = \frac{T(h)}{m}- \frac{\phi(h) S v^2}{2 m} (C_{D1} +C_{D2} u^2) - g sin(\gamma)\\
\frac{dm}{dt} = -C_s(v) T(h)\\
\epsilon\frac{d\gamma}{dt} = \frac{\phi(h) S v}{2 m} u - \frac{g}{v}cos(\gamma)
$$

The Hamiltonien is given by:
$$
 H = p_1 v sin(\gamma) + p_2 (\frac{T(h)}{m} - \frac{\phi(h) S v^2}{2 m}(C_{D1} + C_{D2} u^2 ) -g sin(\gamma)) -p_3 C_s(v) T(h) + \frac{1}{\varepsilon} p_4(\frac{\phi(h) S v}{2 m} u - \frac{g}{v} cos(\gamma) )
 $$

Let x =(h,v,m,$\gamma$), then the hamiltonien is :
$$
 H = p_1 x_2 sin(x_4) + p_2 (\frac{T(x_1)}{x_3} - \frac{\phi(x_1) S x_2^2}{2 x_3}(C_{D1} + C_{D2} u^2 ) -g sin(x_4)) -p_3 C_s(x_2) T(x_1) + \frac{1}{\varepsilon} p_4(\frac{\phi(x_1) S x_2}{2 x_3} u - \frac{g}{x_2} cos(x_4) )
$$

The maximization conditions lead to :
$$
          \dot{p}(t) = -\frac{\partial H}{\partial x} = - \left(\begin{array}{cc} 
p_2[\frac{1}{x_3} \frac{dT}{dx_1} - \frac{S x_2 ^2}{ 2 x_3}(C_{D1} + C_{D2} u ^2 ) \frac{d\phi}{dx_1}] - p_3 C_s(x_2) \frac{dT}{dx_1} + \frac{p_4 S x_2 u}{2 \varepsilon x_3} \frac{d\phi}{dx_1}\\
p_1 sin(x_4) - p_2 \frac{\phi(x_1) S x_2}{x_3} ( C_{D1} + C_{D2} u ^2 ) - p_3 T(x_1) \frac{dC_s}{dx_2} + \frac{p_4 \phi(x_1) S u}{2 \varepsilon x_3} + \frac{p_4 g}{\varepsilon x_2 ^2} cos(x_4) \\
-\frac{p_2 T(x_1)}{x_3 ^2} + \frac{p_2 \phi(x_1) S x_2 ^2}{2 x_3 ^2} (C_{D1} + C_{D2} u^2 ) - \frac{p_4 \phi(x_1) S x_2 u}{2 \varepsilon x_3 ^2} \\
p_1 x_2 cos(x_4) - g p_2 cos(x_4) + \frac{p_4 g}{\varepsilon x_2} sin(x_4)
\end{array}\right)\\ 
\\
               \dot{x}(t) = \frac{\partial H}{\partial p} =    \left(\begin{array}{cc} 
x_2 sin(x4)\\
\frac{T(x_1)}{x_3} - \frac{\phi(x_1) S x_2^2}{2 x_3}(C_{D1} + C_{D2} u^2 ) -g sin(x_4) \\
-C_s(x_2) T(x_1) \\
\frac{1}{\varepsilon} (\frac{\phi(x_1) S x_2}{2 x_3} u - \frac{g}{x_2} cos(x_4) )
\end{array}\right)\\ 
\\
     0 = \frac{\partial H}{\partial u} = - p_2 \frac{\phi(x_1) S x_2^2}{x_3} C_{D2} u + \frac{p_4 \phi(x_1) S x_2}{2 \varepsilon x_3}\\
$$

With :

$$
T(x_1) = C_{T1} (1 - \frac{x_1}{C_{T2}} + x_1 ^2 C_{T_3})\\
C_s(x_2) = C_{s1} (1 + \frac{x_2}{C_{s2}}) \\
\phi(x_1) = \frac{P_0}{R \theta _0 ^ \frac{g}{\beta R}} (\theta_0 - \beta x_1) ^ {\frac{g }{\beta R}  -1} 
$$


$$
\frac{dT}{dx_1} = C_{T1}(\frac{-1}{C_{T2}} + 2 x_1 C_{T3} ) \\
\frac{dC_s}{dx_2} = \frac{C_{s1}}{C_{s2}} \\
\frac{d\phi}{dx_1} = \beta (1 - \frac{g }{\beta R}  ) \frac{P_0}{R \theta _0 ^ \frac{g}{\beta R}} (\theta_0 - \beta x_1) ^ {\frac{g }{\beta R}  -2}
$$

Then : 
$$
u = \frac{p_4}{2 \varepsilon p_2 C_{D2}x_2}
$$

### Associated boundary problem:

$$
    \left\{
    \begin{array}{l}    
        \dot{x_1}(t) = x_2 sin(x4)\\
        \dot{x_2}(t) = \frac{T(x_1)}{x_3} - \frac{\phi(x_1) S x_2^2}{2 x_3}(C_{D1} + C_{D2} u^2 ) -g sin(x_4)\\
        \dot{x_3}(t) = -C_s(x_2) T(x_1)\\
        \dot{x_4}(t) = \frac{1}{\varepsilon} (\frac{\phi(x_1) S x_2}{2 x_3} u - \frac{g}{x_2} cos(x_4) )\\
        \dot{p_1}(t) =-(p_2[\frac{1}{x_3} \frac{dT}{dx_1} - \frac{S x_2 ^2}{ 2 x_3}(C_{D1} + C_{D2} u ^2 ) \frac{d\phi}{dx_1}] - p_3 C_s(x_2) \frac{dT}{dx_1} + \frac{p_4 S x_2 u}{2 \varepsilon x_3} \frac{d\phi}{dx_1})\\
        \dot{p_2}(t) =-( p_1 sin(x_4) - p_2 \frac{\phi(x_1) S x_2}{x_3} ( C_{D1} + C_{D2} u ^2 ) - p_3 T(x_1) \frac{dC_s}{dx_2} + \frac{p_4 \phi(x_1) S u}{2 \varepsilon x_3} + \frac{p_4 g}{\varepsilon x_2 ^2} cos(x_4))\\
        \dot{p_3}(t) = -(-\frac{p_2 T(x_1)}{x_3 ^2} + \frac{p_2 \phi(x_1) S x_2 ^2}{2 x_3 ^2} (C_{D1} + C_{D2} u^2 ) - \frac{p_4 \phi(x_1) S x_2 u}{2 \varepsilon x_3 ^2})\\
        \dot{p_4}(t) = -(p_1 x_2 cos(x_4) - g p_2 cos(x_4) + \frac{p_4 g}{\varepsilon x_2} sin(x_4))\\
        x(0) = (3480.,151.67,69000.,)\\
        x(t_f) = (9144.,191.,68100.,)
    \end{array}
    \right.
$$

### Pas uniforme:

```julia
function nlpsolve_U(eps, N)
    
    sys = Model(optimizer_with_attributes(Ipopt.Optimizer, "print_level" => 1))
    set_optimizer_attribute(sys,"tol",1e-8)
    set_optimizer_attribute(sys,"constr_viol_tol",1e-6)

    # Parameters
    t0 = 0. 
    
    @variables(sys, begin
        0.5*h0 ≤ h[1:N+1] ≤ 1.5*hf
        50.    ≤ v[1:N+1] ≤ 300.
        0.9*mf ≤ m[1:N+1] ≤ 1.1*m0
        -0.3   ≤ γ[1:N+1] ≤ 0.3
        0.0    ≤ u[1:N+1] ≤ 1.6 # real constraints
        0.     ≤ Δt
    end)

    set_start_value(Δt, 650.0/N);
        
    for i in 1:N+1
        set_start_value(h[i], (h0+hf)/2.);  
        set_start_value(v[i], (v0+vf)/2.);
        set_start_value(m[i], (m0+mf)/2.);
        set_start_value(γ[i], (γ0+γf)/2.);
        set_start_value(u[i], 0.1);
    end

    # Objective
    @objective(sys, Min, Δt)

    @constraints(sys, begin
        con_h0, h[1]   - h0 == 0    
        con_hf, h[N+1] - hf == 0  
        con_v0, v[1]   - v0 == 0     
        con_vf, v[N+1] - vf == 0
        con_m0, m[1]   - m0 == 0
        con_mf, m[N+1] - mf == 0
        con_g0, γ[1]   - γ0 == 0
        con_gf, γ[N+1] - γf == 0   
    end)

    @NLexpressions(sys, begin
        T[j = 1:N+1],  Ct1*(1.0 - h[j]/Ct2 + h[j]^2 * Ct3)
        Cs[j = 1:N+1], Cs1*(1.0 + v[j]/Cs2)
        Θ[j = 1:N+1],  Θ0 - β*h[j]
        P[j = 1:N+1],  P0*(Θ[j]/Θ0)^(g/(β*R))
        ρ[j = 1:N+1],  P[j]/(R*Θ[j])
    end)
    
    
    @NLexpressions(sys, begin          
         dh[j = 1:N+1],  v[j] *sin(γ[j])
         dv[j = 1:N+1],  T[j]/m[j] - (ρ[j]*S*v[j]^2)/(2.0*m[j])*(Cd1+Cd2*u[j]^2) - g*sin(γ[j])
         dm[j = 1:N+1], -Cs[j]*T[j] 
         dγ[j = 1:N+1], (1.0/eps)*( (ρ[j]*S*v[j]*u[j])/(2.0*m[j]) - (g/v[j])*cos(γ[j]) )
    end)

    @NLconstraints(sys, begin    
        con_dh[j = 1:N], h[j+1] == h[j] + 0.5 * Δt * (dh[j+1] + dh[j]) # h
        con_dv[j = 1:N], v[j+1] == v[j] + 0.5 * Δt * (dv[j+1] + dv[j]) # v
        con_dm[j = 1:N], m[j+1] == m[j] + 0.5 * Δt * (dm[j+1] + dm[j]) # m
        con_dγ[j = 1:N], γ[j+1] == γ[j] + 0.5 * Δt * (dγ[j+1] + dγ[j]) # γ
    end)

    status = optimize!(sys)

    if termination_status(sys) == MOI.OPTIMAL
        conv = "yes"
    elseif  termination_status(sys) == MOI.LOCALLY_SOLVED
        conv = "yes"
    elseif termination_status(sys) == MOI.TIME_LIMIT && has_values(sys)
        conv = "yes"
    else
        conv = "no"
    end

    obj = objective_value(sys)
    
    # Retrieves values 
    t = (0:N).*value(Δt)
    h = value.(h)
    v = value.(v)
    m = value.(m)
    γ = value.(γ)
    u = value.(u)
    
    p = -[ [ dual(con_dh[i]), dual(con_dv[i]), dual(con_dm[i]), dual(con_dγ[i]) ] for i in 1:N ];
    ph = zeros((N));
    pv = zeros((N));
    pm = zeros((N));
    pγ = zeros((N));
    
    for i in 1:N
            ph[i] = p[i][1]
            pv[i] = p[i][2]
            pm[i] = p[i][3]
            pγ[i] = p[i][4]
    end  
       
    return (t, h, v, m, γ, u,ph,pv,pm, pγ,conv,obj)
end;
```

```julia
function nlpsolve_V(eps, N)
    sys = Model(optimizer_with_attributes(Ipopt.Optimizer, "print_level" => 1))
    set_optimizer_attribute(sys,"tol",1e-8)
    set_optimizer_attribute(sys,"constr_viol_tol",1e-6)

    # Parameters
    t0 = 0. 
    α  = 1e-3

    @variables(sys, begin
        0.9*h0 ≤ h[1:N+1] ≤ 1.1*hf
        50. ≤ v[1:N+1] ≤ 300.
        0.9*mf ≤ m[1:N+1] ≤ 1.1*m0
        -0.3 ≤ γ[1:N+1] ≤ 0.3
        -1.0 ≤ u[1:N+1] ≤ 1.0
        0. ≤ Δt
        0. ≤ hi[1:N]
    end)

    set_start_value(Δt, 650.0/N);
    
    for i in 1:N
        set_start_value(hi[i], 650.0/N);
    end
    
    for i in 1:N+1
        set_start_value(h[i], (h0+hf)/2.);  
        set_start_value(v[i], (v0+vf)/2.);
        set_start_value(m[i], (m0+mf)/2.);
        set_start_value(γ[i], (γ0+γf)/2.);
        set_start_value(u[i], 0.1);
    end

    # Objective
    @objective(sys, Min, Δt + α* sum((hi-Δt*ones(N)).^2))


    @constraint(sys, con_h, sum(hi) == Δt*N - t0)
    
    @constraints(sys, begin
        con_h0, h[1]   - h0 == 0    
        con_hf, h[N+1] - hf == 0  
        con_v0, v[1]   - v0 == 0     
        con_vf, v[N+1] - vf == 0
        con_m0, m[1]   - m0 == 0
        con_mf, m[N+1] - mf == 0
        con_g0, γ[1]   - γ0 == 0
        con_gf, γ[N+1] - 0.0 == 0   
    end)

    @NLexpressions(sys, begin
        T[j = 1:N+1], Ct1*(1.0 - h[j]/Ct2 + h[j]^2 * Ct3)
        Cs[j = 1:N+1], Cs1*(1.0 + v[j]/Cs2)
        Θ[j = 1:N+1], Θ0 - β*h[j]
        P[j = 1:N+1], P0*(Θ[j]/Θ0)^(g/(β*R))
        ρ[j = 1:N+1], P[j]/(R*Θ[j])
    end)
    
    @NLexpressions(sys, begin
        dT[j = 1:N+1], Ct1*(-1/Ct2 + 2*h[j]*Ct3)
        dCs[j = 1:N+1], Cs1/Cs2
        dρ[j = 1:N+1], β * (1 - g/(β*R)) * P0 * Θ[j]^(g/(β*R) - 2) / (R * Θ0 ^ (g/(β*R)))
    end)
    
    @NLexpressions(sys, begin          
         dh[j = 1:N+1],  v[j] *sin(γ[j])
         dv[j = 1:N+1],  T[j]/m[j] - (ρ[j]*S*v[j]^2)/(2.0*m[j])*(Cd1+Cd2*u[j]^2) - g*sin(γ[j])
         dm[j = 1:N+1], -Cs[j]*T[j] 
         dγ[j = 1:N+1], (1.0/eps)*( (ρ[j]*S*v[j]*u[j])/(2.0*m[j]) - (g/v[j])*cos(γ[j]))
    end)

    @NLconstraints(sys, begin    
        con_dh[j = 1:N], h[j+1] == h[j] + 0.5 * hi[j] * (dh[j+1] + dh[j]) # h
        con_dv[j = 1:N], v[j+1] == v[j] + 0.5 * hi[j] * (dv[j+1] + dv[j]) # v
        con_dm[j = 1:N], m[j+1] == m[j] + 0.5 * hi[j] * (dm[j+1] + dm[j]) # m
        con_dγ[j = 1:N], γ[j+1] == γ[j] + 0.5 * hi[j] * (dγ[j+1] + dγ[j]) # γ
    end)

    status = optimize!(sys)

    if termination_status(sys) == MOI.OPTIMAL
        conv = "yes"
    elseif  termination_status(sys) == MOI.LOCALLY_SOLVED
        conv = "yes"
    elseif termination_status(sys) == MOI.TIME_LIMIT && has_values(sys)
        conv = "yes"
    else
        conv = "no"
    end
    
    obj = objective_value(sys)
    
    # Retrieves values 
    t = (0:N).*value(Δt)
    h = value.(h)
    v = value.(v)
    m = value.(m)
    γ = value.(γ)
    u = value.(u)
    hi = value.(hi)
    p = -[ [ dual(con_dh[i]), dual(con_dv[i]), dual(con_dm[i]), dual(con_dγ[i]) ] for i in 1:N ];
    ph = zeros((N));
    pv = zeros((N));
    pm = zeros((N));
    pγ = zeros((N));
    for i in 1:N
            ph[i] = p[i][1]
            pv[i] = p[i][2]
            pm[i] = p[i][3]
            pγ[i] = p[i][4]
    end  
        
    return (t, h, v, m, γ, u,ph,pv,pm, pγ,hi, conv, obj)
end;
```

```julia
#Création d'une liste des expériences
XP = []
function listeXP(eps_init,pas,eps_end,algo,schema,condI,N)
   for eps in eps_init:pas:eps_end
      append!(XP,[[algo,schema,condI,eps,N]])
   end
   return XP
end
```

```julia
#Exécuter une expérience
function execute(algo,eps,N,condI)
   #n = length(algo(t_f,N))
   #conv= algo(eps,N)[n-1]
   #obj= algo(eps,N)[n]
   #time = @elapsed algo(eps,N)
   time =@elapsed (a = algo(eps,N))
   n = length(a)
   conv = a[n-1]
   obj = a[n] 
   #return conv,obj,time
   return conv,obj,time
end

#Vérifier si une XP est sauvegardée 
function save(file,algo,schema,condI,eps,N)
    
       #Vérifier si le fichier existe ou pas
       if (isfile(file) && (length(readlines(file)) > 1))
           #Si le fichier existe on crée une dataFrame à partir de ce fichier
            #CSV to DataFrame
            df = DataFrame(CSV.File(file))   
       else
            #Création d'une DataFrame vide si le fichier n'existe pas
            df = DataFrame(Algorithm=Any[], Scheme=String[],Initialization=Any[] ,eps=Int[], N=Int[],Convergence = Any[], Objective_Value=Any[], Time_cpu=Any[])
            #DataFrame to CSV
            CSV.write(file,df)           
       end
        exp1 = [];
        #Extraction de trois colonnes du fichier csv et les enregistrer dans une liste
        e = CSV.File(file; select=(i, nm) -> i in ( 2, 4, 5,7))
            for i in 1:length(e)
            for j in 1:4
               append!(exp1,[e[i][j]])
            end
        end
    
        #Vérifier si une expérience est déjà sauvegardée ou pas
        conv,obj,time = execute(algo,eps,N,condI)
        if (issubset([schema,eps,N,obj], exp1)) == false
          #conv,obj,time = execute(algo,t_f,N,condI)
          push!(df,(String(string(algo)),schema,string(condI),eps,N,String(string(conv)),obj,time))
          CSV.write(file,df)
        end
        return exp1
       
end

#Remplissage de la DataFrame et sauvegarde dans le fichier csv
function remp(file,liste)
    for k in 1:length(liste)
      ll = liste[k][:]
      exp1 = save(file,ll[1],ll[2],ll[3],ll[4],ll[5])
    end
    df = DataFrame(CSV.File(file))
    println(df)
    return df
end
```

```julia
function supprimerXP(file,algo,eps,N,condI,schema)
    lignes = readlines(file)
    conv,obj,time = execute(algo,eps,N,condI)
    df = DataFrame(CSV.File(file))
    n = length(lignes)
    index = [k for k in 1:n-1]
    for i in 2:n
        taille = length(split(lignes[i],","))
        a,z,e,r,t,y,u,w,l,p,q,s,d=split(lignes[i],",")

        if ( a ==String(string(algo)) && z == schema  && parse(Int64,l) ==eps && parse(Int64,p) ==N  && q ==String(string(conv)) && parse(Float64,s) ==obj)
            j = i-1
            liste = deleteat!(index, findall(x->x==j,index))
            df1 = df[liste,:]
            println(df1)
            CSV.write(file,df1)
        end
          
    end
    
end
```
